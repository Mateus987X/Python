#### 1.1 - Identificando Valores Ausentes

```python
def func_percent_valores_ausentes(df):
    totalCells = np.product(df.shape)
    
    missingCount = df.isnull().sum()

    totalMissing = missingCount.sum()

    print("O dataset tem", round(((totalMissing/totalCells) * 100), 2), '%', 'de valores ausentes.')
```

```python
func_percent_valores_ausentes(df_dsa)
```

```python
def func_percent_valores_ausentes_coluna(df):
    #Valores Ausentes
    mis_val = df.isnull().sum()
    
    #Porcentagem de valores ausentes
    mis_val_percent = 100 * mis_val / len(df)
    
    #Tipo de dados das colunas com valores ausentes
    mis_val_dtype = df.dtypes

    #Cria um tabela com os resultados
    mis_val_table = pd.concat([mis_val, mis_val_percent, mis_val_dtype], axis = 1)

    #Renomear as colunas
    mis_val_table_renomeada = mis_val_table.rename(
    columns = {0: 'Valores ausentes', 1: '% de Valores Ausentes', 2: 'Dtype'})

    #Classifica a tabela por porcentagem de valores ausentes de forma decrescente e remove colunas sem valores  faltantes
    mis_val_table_renomeada = mis_val_table_renomeada[mis_val_table_renomeada.iloc[:,0] != 0].sort_values('% de Valores Ausentes', ascending = False).round(2)

    print("O dataset tem " + str(df.shape[1]) + " colunas.\n"
        "Encontrado: " + str(mis_val_table_renomeada.shape[0]) + " colunas que tÃªm valores ausentes.")

    if mis_val_table_renomeada.shape[0] == 0:
        return

    return mis_val_table_renomeada
```

```python
mis_val = df_dsa.isnull().sum()
mis_val_percent = 100 * mis_val / len(df_dsa)
mis_val_dtype = df_dsa.dtypes
```

```python
table = pd.concat([mis_val, mis_val_percent, mis_val_dtype], axis = 1)
table.rename(columns = {0: "Valores ausentes", 1: '%', 2: 'Dtype'}, inplace = True)
```

```python
df_missing = func_percent_valores_ausentes_coluna(df_dsa)
```

### 1.3 ImputaÃ§Ã£o com Preenchimento Reverso

```python
# ImputaÃ§Ã£o de valores ausentes usando backward fill
# method = 'bfill': Bfill ou backward-fill propaga o primeiro valor nÃ£o nulo observado para trÃ¡s atÃ© que 
# outro valor nÃ£o nulo seja encontrado.
def func_dsa_bfill(df, col):
    count = df[col].isna().sum()

    df[col] = df[col].fillna(method = 'bfill')

    print(f'{count} valores ausentes na coluna {col} foram substituÃ­dos usando o mÃ©todo de preenchimento reverso.')
```

```python
func_dsa_bfill(df_dsa_limpo, "TCP UL Retrans. Vol (Bytes)")
```

##### Avg RTT DL (ms) e Avg RTT UL (ms) tÃªm as prÃ³ximas porcentagens mais altas de valores ausentes com cerca de 18,% cada. Vamos verificar se as variÃ¡veis estÃ£o enviesadas (nÃ£o seguem uma distribuiÃ§Ã£o normal) usando o mÃ©todo skw(), que retorna o coeficiente de assimetria.

```python
#Analisando com mÃ©todo skew()
df_dsa_limpo['Avg RTT DL (ms)'].skew(skipna = True)
```

```python
# Analisando com Shapiro-wilk
from scipy import stats

#Extraindo a sÃ©rie "AVG"
AVG = df_dsa_limpo['Avg RTT DL (ms)'].dropna()

#Aplique o teste de Shapiro-Wilk
stat, p_valor = stats.shapiro(AVG)

print(f'EstatÃ­stica de teste: {stat}')
print(f'Valor-p: {p_valor}')

# Verifique a hipÃ³tese nula com base no p-valor
alfa = 0.05 # NÃ­vel de significÃ¢ncia

if p_valor > alfa:
    print("NÃ£o hÃ¡ evidÃªncias para rejeitar a hipÃ³tese nula (os dados parecem seguir uma distribuiÃ§Ã£o normal).")
else:
    print("A hipÃ³tese nula Ã© rejeitada (os dados nÃ£o seguem uma distribuiÃ§Ã£o normal).")
```

```python
df_dsa_limpo['Avg RTT UL (ms)'].skew(skipna = True)
```

```python
# # Analisando com Shapiro-wilk
from scipy import stats

#Extraindo a sÃ©rie "AVG"
AVG = df_dsa_limpo['Avg RTT UL (ms)'].dropna()

#Aplique o teste de Shapiro-Wilk
stat, p_valor = stats.shapiro(AVG)

print(f'EstatÃ­stica de teste: {stat}')
print(f'Valor-p: {p_valor}')

# Verifique a hipÃ³tese nula com base no p-valor
alfa = 0.05 # NÃ­vel de significÃ¢ncia

if p_valor > alfa:
    print("NÃ£o hÃ¡ evidÃªncias para rejeitar a hipÃ³tese nula (os dados parecem seguir uma distribuiÃ§Ã£o normal).")
else:
    print("A hipÃ³tese nula Ã© rejeitada (os dados nÃ£o seguem uma distribuiÃ§Ã£o normal).")
```

#### Se a simetria estiver entre -0,5 e 0,5, os dados sÃ£o bastante simÃ©tricos
#### Se a assimetria estiver entre -1 e -0,5 ou entre 0,5 e 1, os dados estÃ£o moderadamente inclinados
#### Se a assimetria for menor que -1 ou maior que 1, os dados estÃ£o altamente enviesados

#### Visto que ambas as colunas sÃ£o fortemente enviesadas positivamente Ã© aconselhÃ¡vel nÃ£o imputÃ¡-las com sua mÃ©dia.
#### Portanto, usaremos o preenchimento progressivo.

```python
def func_dsa_ffil(df, col):

    count = df[col].isna().sum()

    df[col] = df[col].fillna(method = 'ffill')

    print(f'{count} valores ausentes na coluna {col} foram substituÃ­dos usando o mÃ©todo de preenchimento progressivo.')
```

```python
func_dsa_ffil(df_dsa_limpo, 'Avg RTT DL (ms)')
```

```python
func_dsa_ffil(df_dsa_limpo, 'Avg RTT UL (ms)')
```

#### Visto que "Handset Type" e "Handset Manufacturer" sÃ£o colunas categÃ³ricas, Ã© melhor imputÃ¡-las com o valor "unknown" para que nÃ£o enviesemos os dados

```python
# Preenche valor NA
def func_preencher_valor_NA_categorico(df, col, value):
    count = df[col].isna().sum()

    df[col] = df[col].fillna(value)

    if type(value) == 'str':
        print(f'{count} valores ausentes na coluna {col} foram substituÃ­dos por "{value}".')
    else:
        print(f'{count} valores ausentes na coluna {col} foram substituÃ­dos por {value}.')
```

```python
func_preencher_valor_NA_categorico(df_dsa_limpo, 'Handset Type', 'unknown')
func_preencher_valor_NA_categorico(df_dsa_limpo, 'Handset Manufacturer', 'unknown')
```

```python
def func_dsa_drop_linhas_com_na(df):

    old = df.shape[0]

    df.dropna(inplace = True)

    new = df.shape[0]

    count = old - new

    print(f'{count} linhas contendo valores ausentes foram descartadas.')
```

```python
func_dsa_drop_linhas_com_na(df_dsa_limpo)
```
