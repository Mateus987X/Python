```python
# Esse programa converte um arquivo Excel para o formato DBF.

import pandas as pd
import dbf

def excel_to_dbf(excel_path, dbf_path, sheet_name=0):
    # Ler o arquivo Excel
    df = pd.read_excel(excel_path, sheet_name=sheet_name)
    
    # Criar estrutura da tabela DBF
    fields = []
    for col_name, col_type in zip(df.columns, df.dtypes):
        if col_type in ['int64', 'int32']:
            field_def = f"{col_name} N(10,0)"  # Campo numérico inteiro
        elif col_type in ['float64', 'float32']:
            field_def = f"{col_name} N(15,4)"  # Campo numérico com decimais
        else:
            # Para campos de texto, limitamos a 254 caracteres (limitação do DBF)
            field_def = f"{col_name} C(254)"
        
        fields.append(field_def)
    
    # Criar tabela DBF
    table = dbf.Table(dbf_path, fields)
    table.open(dbf.READ_WRITE)
    
    # Adicionar registros
    for _, row in df.iterrows():
        record = []
        for col, col_type in zip(df.columns, df.dtypes):
            value = row[col]
            
            if pd.isna(value):
                record.append(None)
            elif col_type in ['int64', 'int32']:
                record.append(int(value))
            elif col_type in ['float64', 'float32']:
                record.append(float(value))
            else:
                # Converter para string e truncar se necessário
                str_value = str(value)[:254] if value else ''
                record.append(str_value)
        
        table.append(tuple(record))
    
    table.close()
    print(f"Arquivo DBF criado com sucesso: {dbf_path}")
```

```python
excel_to_dbf('SA1SEG.xlsx', 'saida.dbf')
```
